```python
df = pandas.read_csv( 'raw_collect.txt', sep='\s+', header=None, names=['mutant', 'student'] )
allowed = pandas.read_csv( '../../../../bagel-data/allowed_mutants.txt', header=None, names=['mutant'] )
inhouse = pandas.read_csv( '../../../../bagel-benchmark/mutant_list', header=None, names=['mutant'] )

m_list = allowed.mutant.tolist()
in_list = inhouse.mutant.tolist()

print 'name\tallowed\tin-house'
for mut in df.mutant:
    print mut, '\t', mut in m_list, '\t', mut in in_list
```

```python
import screed 

with open( 'csv.csv', 'w' ) as fn:
    fn.write( 'mutant_label,oligo_label,sequence,scale,purification\n' )
    for line in df.mutant.unique():
        for record in screed.open( '../../../oligos/{}.fasta'.format( line ) ):
            fn.write( '{0},{0},{1},25nm,standard\n'.format( line, record.sequence ) )
```
